```python
# Mutable data types
# Can be changed after creation (modified in place).
# Examples: list, dict, set
```

```python
a = [1, 2, 3]
a.append(4)   # a becomes [1, 2, 3, 4]
print(a)
```

```python
# Immutable data types
# Cannot be changed after creation (a new object is created instead).
# Examples: int, float, str, tuple
```

```python
t = (1, 2, 3)
t[0] = 10   

```
